```python
## 1. Setup and Imports

import numpy as np
import matplotlib.pyplot as plt

# Set plot style
plt.style.use('seaborn-v0_8-darkgrid')
```

```python
## 2. Define Economic Functions

def demand(price, a=100, b=2):
    """Demand: Qd = a - b*P"""
    return a - b * price

def supply(price, c=10, d=1.5):
    """Supply: Qs = c + d*P"""
    return c + d * price

def find_equilibrium(a, b, c, d):
    """Calculate equilibrium price and quantity"""
    p_eq = (a - c) / (b + d)
    q_eq = demand(p_eq, a, b)
    return p_eq, q_eq
```

```python
## 3. Basic Market Equilibrium

prices = np.linspace(0, 60, 300)
qd = demand(prices)
qs = supply(prices)
p_eq, q_eq = find_equilibrium(100, 2, 10, 1.5)

plt.figure(figsize=(10, 6))
plt.plot(qd, prices, 'b-', linewidth=2, label='Demand')
plt.plot(qs, prices, 'r-', linewidth=2, label='Supply')
plt.plot(q_eq, p_eq, 'go', markersize=12, label=f'Equilibrium (Q={q_eq:.1f}, P=${p_eq:.1f})')
plt.axhline(y=p_eq, color='gray', linestyle='--', alpha=0.5)
plt.axvline(x=q_eq, color='gray', linestyle='--', alpha=0.5)

plt.xlabel('Quantity', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.title('Market Equilibrium', fontsize=14, fontweight='bold')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 100)
plt.ylim(0, 60)
plt.tight_layout()
plt.show()

print(f"Equilibrium Price: ${p_eq:.2f}")
print(f"Equilibrium Quantity: {q_eq:.2f} units")
```

```python
## 4. Market Shifts

### 4.1 Demand Increase

# Increased demand (a = 120 instead of 100)
qd_new = demand(prices, a=120)
p_eq_new, q_eq_new = find_equilibrium(120, 2, 10, 1.5)

plt.figure(figsize=(10, 6))
plt.plot(qd, prices, 'b--', linewidth=1.5, alpha=0.5, label='Original Demand')
plt.plot(qd_new, prices, 'b-', linewidth=2, label='New Demand')
plt.plot(qs, prices, 'r-', linewidth=2, label='Supply')
plt.plot(q_eq, p_eq, 'go', markersize=10, alpha=0.5, label=f'Old Equilibrium')
plt.plot(q_eq_new, p_eq_new, 'go', markersize=12, label=f'New Equilibrium')

plt.xlabel('Quantity')
plt.ylabel('Price ($)')
plt.title('Effect of Demand Increase')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 100)
plt.ylim(0, 60)
plt.tight_layout()
plt.show()

print(f"Price increased from ${p_eq:.2f} to ${p_eq_new:.2f}")
print(f"Quantity increased from {q_eq:.2f} to {q_eq_new:.2f}")
```

```python
### 4.2 Supply Increase

# Increased supply 
qs_new = supply(prices, c=25)
p_eq_new2, q_eq_new2 = find_equilibrium(100, 2, 25, 1.5)

plt.figure(figsize=(10, 6))
plt.plot(qd, prices, 'b-', linewidth=2, label='Demand')
plt.plot(qs, prices, 'r--', linewidth=1.5, alpha=0.5, label='Original Supply')
plt.plot(qs_new, prices, 'r-', linewidth=2, label='New Supply')
plt.plot(q_eq, p_eq, 'go', markersize=10, alpha=0.5, label=f'Old Equilibrium')
plt.plot(q_eq_new2, p_eq_new2, 'go', markersize=12, label=f'New Equilibrium')

plt.xlabel('Quantity')
plt.ylabel('Price ($)')
plt.title('Effect of Supply Increase')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 100)
plt.ylim(0, 60)
plt.tight_layout()
plt.show()

print(f"Price decreased from ${p_eq:.2f} to ${p_eq_new2:.2f}")
print(f"Quantity increased from {q_eq:.2f} to {q_eq_new2:.2f}")
```

```python
## 5. Government Interventions

### 5.1 Price Floor (e.g., Minimum Wage)

floor_price = 30
qd_floor = demand(floor_price)
qs_floor = supply(floor_price)
surplus = qs_floor - qd_floor

plt.figure(figsize=(10, 6))
plt.plot(qd, prices, 'b-', linewidth=2, label='Demand')
plt.plot(qs, prices, 'r-', linewidth=2, label='Supply')
plt.plot(q_eq, p_eq, 'go', markersize=10, label=f'Equilibrium')
plt.axhline(y=floor_price, color='orange', linewidth=2.5, label=f'Price Floor (${floor_price})')
plt.fill_betweenx([floor_price, floor_price], qd_floor, qs_floor, alpha=0.3, color='yellow')

plt.xlabel('Quantity')
plt.ylabel('Price ($)')
plt.title('Price Floor Creates Surplus')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 100)
plt.ylim(0, 60)
plt.tight_layout()
plt.show()

print(f"Surplus created: {surplus:.2f} units")
print(f"Quantity demanded: {qd_floor:.2f}, Quantity supplied: {qs_floor:.2f}")
```

```python
### 5.2 Price Ceiling (e.g., Rent Control)

ceiling_price = 20
qd_ceiling = demand(ceiling_price)
qs_ceiling = supply(ceiling_price)
shortage = qd_ceiling - qs_ceiling

plt.figure(figsize=(10, 6))
plt.plot(qd, prices, 'b-', linewidth=2, label='Demand')
plt.plot(qs, prices, 'r-', linewidth=2, label='Supply')
plt.plot(q_eq, p_eq, 'go', markersize=10, label=f'Equilibrium')
plt.axhline(y=ceiling_price, color='purple', linewidth=2.5, label=f'Price Ceiling (${ceiling_price})')
plt.fill_betweenx([ceiling_price, ceiling_price], qs_ceiling, qd_ceiling, alpha=0.3, color='pink')

plt.xlabel('Quantity')
plt.ylabel('Price ($)')
plt.title('Price Ceiling Creates Shortage')
plt.legend()
plt.grid(True, alpha=0.3)
plt.xlim(0, 100)
plt.ylim(0, 60)
plt.tight_layout()
plt.show()

print(f"Shortage created: {shortage:.2f} units")
print(f"Quantity demanded: {qd_ceiling:.2f}, Quantity supplied: {qs_ceiling:.2f}")
```

```python
## 6. Elasticity Analysis

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 5))

# Inelastic demand (steep)
qd_inelastic = demand(prices, a=100, b=0.5)
ax1.plot(qd_inelastic, prices, 'b-', linewidth=2, label='Inelastic Demand')
ax1.plot(qs, prices, 'r-', linewidth=2, label='Supply')
ax1.set_xlabel('Quantity')
ax1.set_ylabel('Price ($)')
ax1.set_title('Inelastic Demand (Necessities)')
ax1.legend()
ax1.grid(True, alpha=0.3)
ax1.set_xlim(0, 120)
ax1.set_ylim(0, 60)

# Elastic demand (flat)
qd_elastic = demand(prices, a=100, b=3)
ax2.plot(qd_elastic, prices, 'b-', linewidth=2, label='Elastic Demand')
ax2.plot(qs, prices, 'r-', linewidth=2, label='Supply')
ax2.set_xlabel('Quantity')
ax2.set_ylabel('Price ($)')
ax2.set_title('Elastic Demand (Luxuries)')
ax2.legend()
ax2.grid(True, alpha=0.3)
ax2.set_xlim(0, 120)
ax2.set_ylim(0, 60)

plt.tight_layout()
plt.show()

print("Inelastic: Quantity less responsive to price changes (e.g., insulin, gasoline)")
print("Elastic: Quantity very responsive to price changes (e.g., luxury goods, vacations)")
```

## Conclusion

This analysis demonstrates:
- **Market equilibrium** occurs where supply equals demand
- **Demand increases** raise both price and quantity
- **Supply increases** lower price but raise quantity  
- **Price floors** create surpluses (unemployment in labor markets)
- **Price ceilings** create shortages (housing shortages with rent control)
- **Elasticity** determines how sensitive quantity is to price changes
